#I am going to compute the average time for the trip based on the starting point.

```python
%%bigquery --project bustling-cosmos-338722
SELECT 
  start_station_name, 
  AVG(duration_minutes) as Average_time_for_trip
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name is not null 
GROUP BY start_station_name
```

#Here, I will compute exactly how many trip starts at each of the starting points. 

```python
%%bigquery --project bustling-cosmos-338722
SELECT 
  start_station_name,
  COUNT(*) as total_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY total_trips DESC
```
