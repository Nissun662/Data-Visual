#Today's Assignment: 
Grab the `iris` dataset.  Answer the following questions:

1. Does converting *SepalLength* to integer increase or decrease the mean?
2. Does the direction of the shift remain the same if you `groupby` Class?
3. Gather the mean, median, count and standard deviation of all columns when grouped by Class.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')
df.head(10)
```

```python
df.SepalLength.dtypes
```

SepalLengh mean in `Float` datatype.

```python
df.SepalLength.mean()
```

Converting the SepalLenght into `Integer` datatype and finding its mean.

```python
df.SepalLength.astype('int').mean()
```

By looking at the above two output we can cleary say mean decreased after converting it into `int` dtypes.

Grouping the iris dataset by Class.

```python
df.groupby('Class')
df.head(10)
```

After I grouped the iris dataset by `Class` nothing changed so I would assume there is no change in the direction of shift.

## Finding the mean, median, count and standard deviation after grouping by Class.

```python
df.groupby('Class').agg(['mean','median','count','std'])
```
