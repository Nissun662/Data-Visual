Now, I will use selenium to browse the webpage I want. I am also going to import `By` and `Keys` from selenium.

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 

url = "https://www.google.com"
driver.get(url)
element = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
element.send_keys('WWE')
element.send_keys(Keys.ENTER)
```

Here, the google search page for any given keyword appears. Currently I am using WWE as the keyword.

```python
driver.current_url
```

Navigating to the very first result and opening the link.

```python
result = driver.find_element(By.CLASS_NAME, 'yuRUbf')
result.click()
```
