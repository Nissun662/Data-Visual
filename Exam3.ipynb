##(c) (10 points) Create a histogram of the ‘Goals’.

The histogram of total goals scored by Argentina.

```python
ax = df.Goals.hist(bins = 15)
ax.set(title = "Histogram of Goals by Argentina")
ax.set_xlabel("Total Goals", labelpad=15, size=10)
ax.set_ylabel("Players", labelpad=15, size=10)
plt.show()
```

##(d) (10 points) Create a histogram of the ‘Goals’ and ‘Assists’ on the same axis.

```python
df[['Goals','Assists']].plot.hist(alpha = 0.8, bins = 50)
```

##(e) (10 points) Create a barchart for most goals scored by nationality.

```python
ax = df.groupby('nationality').Goals.agg('mean').plot(kind = 'bar', rot =85)
ax.set(title = 'Most Goals by Nationality')
ax.set(ylabel = 'Total Goals')
plt.show()
```

```python
ax = df.groupby('nationality').Goals.agg('mean').plot(kind = 'barh')
ax.set(title = 'Most goals scored by Nationality')
ax.set(ylabel = 'Goals')
plt.show()
```
