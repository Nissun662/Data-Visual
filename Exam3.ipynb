##(c) (10 points) Create a histogram of the ‘Goals’.

The histogram of total goals scored by Argentina.

```python
ax = df.Goals.hist(bins = 15)
ax.set(title = "Histogram of Goals by Argentina")
ax.set_xlabel("Total Goals", labelpad=25, size=15)
ax.set_ylabel("Players", labelpad=25, size=15)
plt.show()
```

##(d) (10 points) Create a histogram of the ‘Goals’ and ‘Assists’ on the same axis.

```python
df1=df[['Goals','Assists']].plot.hist(title='Histogram of Goals and Assist', bins=10)
df1.set_xlabel('Goals')
df1.set_ylabel('Players')
```

##(e) (10 points) Create a barchart for most goals scored by nationality.

```python
ax = df.groupby('nationality').Goals.agg('mean').plot(kind = 'bar', rot =85)
ax.set(title = 'Most Goals by Nationality')
ax.set(ylabel = 'Total Goals')
plt.show()
```

```python
ax = df.groupby('nationality').Goals.agg('mean').plot(kind = 'barh', rot=12)
ax.set(title = 'Most goals scored by Nationality')
ax.set(ylabel = 'Goals')
plt.show()
```

##(f) (10 points) Create a boxplot of ‘Yel’ (yellow cards) by nationality.

```python
df.boxplot(by='nationality', column='Yel', rot=90).set_ylabel('Yellow card')
```

##(g) (10 points) Create ‘Goals’ and ‘Assists’ as scatter. Signify your favorite club or nationality in some way on the scatterplot. You may need to add a new column to the dataset to do so.

```python
ax=sns.scatterplot(data = df, 
                   y = 'Goals', 
                   x = 'Assists', 
                   hue = (df['nationality']=='Argentina'))
ax.set_title('Argentina Goals and Assists')
plt.show()
```

```python
ax = sns.scatterplot(data = df,  x = 'Goals', y = 'Assists', style = 'Goals', size = 'Assists' ).set_title('Assist and Goal by Argentina')
```
