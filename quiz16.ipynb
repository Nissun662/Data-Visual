```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/IOT-temp.csv')

df.head(15)
```

```python
df.describe()
```

1. Converting the noted_date into DateTime i.e; Day-month-year H:M style.

```python
pd.to_datetime(df.noted_date,format='%d-%m-%Y %H:%M')
```

As we can see in the above output, noted_date is now converted into `datetime64` datatype format. Noe, I am going to create new column for it which I am going to name DateTime. 

```python
df = df.assign(DateTime = pd.to_datetime(df.noted_date,format='%d-%m-%Y %H:%M'))
df
```

Calculating the maximum and minimum date and time:

```python
print('maximum date and time: ' + str(max(df.DateTime)))

print('minimum date and time: ' + str(min(df.DateTime)))
```

Calculating the largest interval inbetween the temperature checks:

```python
df['DateTime'].max() - df['DateTime'].min()
```

To calculate average datetime:

```python
df.DateTime.mean()
```
