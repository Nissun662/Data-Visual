#Today's Assignment: Nan's and Imputing 

Check out the Air B&B dataset, https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv.  Examine how many entries are null.  Impute for atleast two columns that are null in an approriate fashion.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.head()
```

To computing the number of NULL entries in the dataset.

```python
df.isnull().sum(axis = 0)
```

```python
df.isnull().sum().sum()
```

There are 20,141 NULL entries in the given dataset.

I'm going to choose `last_review` and `reviews_per_month` to impute it into appropriate datatypes.

```python
df.reviews_per_month.fillna(value=df.reviews_per_month.mean(), inplace=True)
df
```

```python
df.isnull().sum()
```

We can see the null values are gone is the reviews_per_month and replaced.

Now I will check the dtypes of last_review and then change it into appropriate datatype.

```python
df.last_review.dtypes
```

The object dtypes is inappropriate as the information should have been shown in date and time format.Here I will make the changes.

```python
df.last_review = pd.to_datetime(df.last_review)
df
```

```python
df.last_review.dtype
```

Changed from object dtype to Date and time datatype `<M8[ns]`

```python
df.isnull().sum(axis = 0)
```

Hence, I have made appropriate changes to the column last_review and reviews_per_month.
