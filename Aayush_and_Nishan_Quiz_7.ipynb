#Liquor sales in Polk county in Dollar and Volume in Gallons
From 2020-04-01 to 2020-04-30.

```python
%%bigquery --project bustling-cosmos-338722
SELECT county as county, SUM(sale_dollars) as dollar_amt, date as date
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2020-04-01' AND '2020-04-30' AND county = "POLK"
GROUP BY county, date 
LIMIT 5
```

From 2021-04-01 to 2021-04-30

```python
%%bigquery --project bustling-cosmos-338722
SELECT county as county, SUM(sale_dollars) as dollar_amt, date as date
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2021-04-01' AND '2021-04-30' AND county = "POLK"
GROUP BY county, date
LIMIT 5
```

Now, I am going to add all the dollar amount is sales and the volume amount in gallons for the month April, 2020.

```python
%%bigquery --project bustling-cosmos-338722
SELECT county as county, SUM(sale_dollars) as total_sales_amount, SUM(volume_sold_gallons) as volume_in_gallons
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2020-04-01' AND '2020-04-30' AND county = "POLK"
GROUP BY county
```

Similarly, I am going to add all the dollar amount is sales and the volume amount in gallons for the month April, 2021.


```python
%%bigquery --project bustling-cosmos-338722
SELECT county as county, SUM(sale_dollars) as total_sales_amount, SUM(volume_sold_gallons) as volume_sold_gallons
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN '2021-04-01' AND '2021-04-30' AND county = "POLK"
GROUP BY county
```

The above two dataset represent the total amount sold in both dollar amount and Volume in Gallons
