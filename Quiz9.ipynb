First of all, I will load necessary libraries.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

As instructed I will be using the wikipedia link https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films to gather tables.

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films')
html_parse = r.text
html_soup = BeautifulSoup(html_parse,'lxml')
```

```python
len(html_soup.find_all('table'))
```

We can see that there are 7 tables in a list.

Now, we can get the table from the webpage.

```python
marvelTable = html_soup.find_all('table', class_ = 'wikitable plainrowheaders')
dataFrame = pd.read_html(str(marvelTable))[0]
dataFrame
```

Removing unnecessary rows over the heading using droplevel method in pandas.

```python
dataFrame.columns = dataFrame.columns.droplevel(1)
dataFrame
```

Looking at the table I can see row 6 and 13 are not movies so I have to remove those two rows using the same drop method from pandas. As I do that it messes up the order of the rows so I have to fix that too. 

```python
dataFrame = dataFrame.drop(dataFrame.index[[6,13]])
dataFrame
```

```python
dataFrame = dataFrame.reset_index(drop=True, inplace=False)
dataFrame
```

Now, I will remove the column Unnamed: 5_level_0 from the index as it does not reveal kind of information.

```python
del dataFrame['Unnamed: 5_level_0']
dataFrame
```
