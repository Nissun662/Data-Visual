##(e) (10 points) Impute ‘total_steps’ by replacing the ‘NaN’s with an appropriate number of steps. Convert to appropriate datatype.

```python
df.total_steps.fillna(value=df.total_steps.min(), inplace = True)
df.head(10)
```

```python
df.total_steps = df.total_steps.astype(int) ##changed it to integer datatype
df.head(5)
```

##(f) (10 points) Convert non-empty ‘avg_pace’ into a float that is still representative of the information contained in the column. Recall that there are 60 seconds in one minute so 3:30 is equivalent to 3.5 minutes.

= Since, I already checked the dtypes in the dataset and found out that the `avg_page` has an object datatype. I'll run the following command to change it to a float dtype.


```python
df.avg_pace = pd.to_datetime(df.avg_pace)
df.avg_pace.dtypes
```

```python
df.avg_pace = df.avg_pace.apply(lambda avg_pace: avg_pace.hour + (avg_pace.minute / 60))
```
