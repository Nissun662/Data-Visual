```python

import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Nissun662/Data-Visual/main/FastestMammals.csv')

df.head(13)
```

```python
x = df['Mammals']
y = df['MPH']
plt.title('Top 10 Highest-Paid Athlete of All-Time')
ax=plt.barh(x, y)
plt.show()
```

```python

# Plot Graph

plt.barh(x, y)
ax=plt.axes()
plt.barh(x, y, color=('#C04000'), linewidth=5, edgecolor='#EBF4FA')
plt.title('Top 10 Fastest Land Mammals')
plt.xlabel('MPH')
plt.ylabel('Animal')


# Set color

plt.barh(x, y, color=['#C04000'])
ax.set_facecolor('#87CEEB')

for barh, x in zip(ax.patches, x[::1]):
    ax.text(1, barh.get_y()+barh.get_height()/2, x, color = 'white', ha = 'left', va = 'center', fontweight = 'bold')  

# Display Graph

plt.show()
```
