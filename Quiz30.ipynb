```python
#Using NYC air b&b data to create a multiple bar graph with 'neighbourhood_group' and 'room_type' by looking at the average 'price'.
import matplotlib
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.groupby(['neighbourhood_group','room_type']).price.agg('max').plot(kind = 'barh',
                                                 title = 'Max price', 
                                                 ylabel= 'Price in dollar')

```

```python
#Using the workout dataset to create a scatterplot with as many features as possible

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

sns.set(rc={"figure.figsize":(10, 10)})
ax = sns.scatterplot(data = df, 
                x = 'distance', 
                y = 'time',             
                size = 'distance',
                hue = 'workout_type',
                x_bins = "avg_speed")
sns.move_legend(ax, "upper right", bbox_to_anchor=(1,1))

```
